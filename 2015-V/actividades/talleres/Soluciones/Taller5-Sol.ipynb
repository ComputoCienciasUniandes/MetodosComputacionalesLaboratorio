# Taller 5 - Fenómeno de Gibbs

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
from scipy.optimize import minimize
from scipy.optimize import basinhopping
```

```python
def sumcs(x,n):
    counter = 0
    y=0
    for i in range(n):
        y+=1./(2*i+1.)*np.sin((2*i+1)*x)
    return y
```

```python
result=minimize(lambda x: -sumcs(x,3),np.pi,method='Nelder-Mead')
```

```python
x=np.linspace(0,2*np.pi,100)
fig=plt.figure()
plt.plot(x,sumcs(x,10))
plt.show()
```

```python
gibbs=[]
for i in range(2,20):
    result=minimize(lambda x: -sumcs(x,i),np.pi,method='Nelder-Mead')
    gibbs.append([i,-result.fun])
gibbsa=np.array(gibbs)
plt.figure()
plt.plot(gibbsa[:,0],gibbsa[:,1],"o-")
plt.plot(gibbsa[:,0],np.pi/4.+np.pi/2*0.089+np.zeros(len(gibbsa)))
plt.show()
```

```python
basinhopping(lambda x: -sumcs(x,i),np.pi)
```

```python
gibbs=[]
for i in range(2,30,2):
    result=basinhopping(lambda x: -sumcs(x,i),np.pi,niter=200)
    gibbs.append([i,-result.fun])
gibbsa=np.array(gibbs)
plt.figure(figsize=(10,5))
plt.plot(gibbsa[:,0],(gibbsa[:,1]-np.pi/4.)/np.pi*2,"o-")
#plt.plot(gibbsa[:,0],np.pi/4.+np.pi/2*0.089+np.zeros(len(gibbsa)))
plt.title("Constante de Gibbs\n")
plt.tick_params(axis='both',labelsize=12)
plt.xlabel("N",fontsize=15)
plt.ylabel(r"$(G-\pi/2)/\pi$",fontsize=15)
plt.savefig("/Users/juan/Dropbox/gibbs.png")
plt.show()
```
