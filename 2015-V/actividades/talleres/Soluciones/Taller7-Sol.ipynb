```python
def uvec1(theta):
    return np.array([-np.sin(theta),np.cos(theta),0])
def uvec2(theta):
    return np.array([-1./np.sqrt(2)*np.cos(theta),-1./np.sqrt(2.)*np.sin(theta),1./np.sqrt(2.)])
```

```python
times = np.linspace(0,600,20000)
from matplotlib import animation
%matplotlib osx
fig=plt.figure()
ax = fig.add_subplot(111, autoscale_on=False,xlim=(-25., 25.), ylim=(-20,37)) #, 
#ax.grid()
line, = ax.plot([], [], '-', ms=1)
plt.axis('off')
angles=np.linspace(0,2*np.pi,100)
ross_sol = odeint(rossler,Yinicial,times)[-4000:]

def animate(i):
    ang=angles[i]
    proyeccion=np.array([np.apply_along_axis(sum,1,ross_sol*uvec1(ang)),np.apply_along_axis(sum,1,ross_sol*uvec2(ang))])
    proyeccion=proyeccion.transpose()
    thisx = proyeccion[:,0]
    thisy = proyeccion[:,1]
    line.set_data(thisx, thisy)
    return line

ani = animation.FuncAnimation(fig, animate, np.arange(0,100),
    interval=1, blit=False)
#ani.save("/Users/juan/Documents/Trabajo/Andes/2015-V/MetodosComputacionalesLaboratorio/2015-V/actividades/talleres/Taller6/RosslerAnimation.mp4",fps=15,dpi=200)
plt.show()
```
