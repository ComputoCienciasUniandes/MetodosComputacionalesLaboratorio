# Taller 8 - Onda en un Cubo

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
def update(topo,cara,time):
    global r
    global totem
    global Nl
    
    ###### current time
    up=totem[time,topo[0,0]]
    up=np.rot90(up,(topo[0,1]+2)%4)
    
    right=totem[time,topo[1,0]]
    right=np.rot90(right,(topo[1,1]+1)%4)
    
    down=totem[time,topo[2,0]]
    down=np.rot90(down,(topo[2,1])%4)
    
    left=totem[time,topo[3,0]]
    left=np.rot90(left,(topo[3,1]+3)%4)
        
    ###### previous time
    up_p=totem[time-1,topo[0,0]]
    up_p=np.rot90(up_p,(topo[0,1]+2)%4)
    
    right_p=totem[time-1,topo[1,0]]
    right_p=np.rot90(right_p,(topo[1,1]+1)%4)
    
    down_p=totem[time-1,topo[2,0]]
    down_p=np.rot90(down_p,(topo[2,1])%4)
    
    left_p=totem[time-1,topo[3,0]]
    left_p=np.rot90(left_p,(topo[3,1]+3)%4)
    
    ###### previous previous time
    up_pp=totem[time-2,topo[0,0]]
    up_pp=np.rot90(up_pp,(topo[0,1]+2)%4)
    
    right_pp=totem[time-2,topo[1,0]]
    right_pp=np.rot90(right_pp,(topo[1,1]+1)%4)
    
    down_pp=totem[time-2,topo[2,0]]
    down_pp=np.rot90(down_pp,(topo[2,1])%4)
    
    left_pp=totem[time-2,topo[3,0]]
    left_pp=np.rot90(left_pp,(topo[3,1]+3)%4)
    
    ###### Three times for the chosen face
    face=totem[time-2:time+1,cara]
    
    # augment it
    augm=np.zeros((Nl+2,Nl+2))
    augm[1:-1,1:-1]=face[2]
    augm[0,1:-1]=up[-1]
    augm[-1,1:-1]=down[0]
    augm[1:-1,0]=left[:,-1]
    augm[1:-1,-1]=right[:,0]
    
    # augment for the previous time
    augmp=np.zeros((Nl+2,Nl+2))
    augmp[1:-1,1:-1]=face[1]
    augmp[0,1:-1]=up_p[-1]
    augmp[-1,1:-1]=down_p[0]
    augmp[1:-1,0]=left_p[:,-1]
    augmp[1:-1,-1]=right_p[:,0]
    
    # augmented for the one before that one
    augmpp=np.zeros((Nl+2,Nl+2))
    augmpp[1:-1,1:-1]=face[0]
    augmpp[0,1:-1]=up_pp[-1]
    augmpp[-1,1:-1]=down_pp[0]
    augmpp[1:-1,0]=left_pp[:,-1]
    augmpp[1:-1,-1]=right_pp[:,0]
    
    # do wave equation
    new=2*(1-2.*r**2)*augmp-augmpp+r**2*(np.roll(augmp,1,axis=0)+np.roll(augmp,-1,axis=0)+np.roll(augmp,1,axis=1)+np.roll(augmp,-1,axis=1))
    
    # return with the ficticious borders chopped
    return new[1:-1,1:-1]
```

```python
# The topology defines the way the sides of all faces are connected
# up is 0
# right is 1
# bottom is 2
# left is 3

# faces   4
#       0 1 2 3
#         5

topology=np.array([[[4, 3],[1, 3],[5, 3],[3, 1]],
                   [[4, 2],[2, 3],[5, 0],[0, 1]],
                   [[4, 1],[3, 3],[5, 1],[1, 1]],
                   [[4, 0],[0, 3],[5, 2],[2, 1]],
                   [[3, 0],[2, 0],[1, 0],[0, 0]],
                   [[1, 2],[2, 2],[3, 2],[0, 2]]])
```

## All bumped

```python
Nt=2000
faces=6
Nl=50
L=10. # Lado del cubo
tmin=0
tmax=.3
xmin=-L/2
xmax=L/2
ymin=-L/2
ymax=L/2
dt=(tmax-tmin)/Nt
dx=(xmax-xmin)/Nl
v=425. # m/s
r=v*dt/dx
print r
xcoords=np.linspace(xmin,xmax,Nl)
ycoords=np.linspace(xmin,xmax,Nl)
xmesh,ymesh=np.meshgrid(xcoords,ycoords)
```

```python
totem=np.zeros((Nt,faces,Nl,Nl))
# fijar la condición inicial
# bumps in all faces
for i in range(6):
    totem[0,i]=np.exp(-xmesh**2)*np.exp(-ymesh**2)
    #totem[0,1][xmesh**2+ymesh**2>16.]=0
    # en la primera iteración no se afectan otras caras
    totem[1,i]=totem[0,i]+r**2/2.*(np.roll(totem[0,i],1,axis=0)+np.roll(totem[0,i],-1,axis=0)+np.roll(totem[0,i],1,axis=1)+np.roll(totem[0,i],-1,axis=1)-4*totem[0,i])

# iterar    
for i in range(2,Nt):
    temptotem=np.zeros((faces,Nl,Nl))
    for j in range(6):
        temptotem[j]=update(topology[j],j,i)
    for j in range(6):
        totem[i,j]=temptotem[j]
```

```python
%matplotlib osx
chosentotem=totem[i]
supertotem=np.zeros((3*Nl,4*Nl))
supertotem[0:Nl,Nl:2*Nl]=chosentotem[4]
supertotem[Nl:2*Nl,Nl:2*Nl]=chosentotem[1]
supertotem[2*Nl:3*Nl,Nl:2*Nl]=chosentotem[5]
supertotem[Nl:2*Nl,0:Nl]=chosentotem[0]
supertotem[Nl:2*Nl,2*Nl:3*Nl]=chosentotem[2]
supertotem[Nl:2*Nl,3*Nl:4*Nl]=chosentotem[3]

fig=plt.figure()
tplot=plt.imshow(supertotem,cmap='jet',vmin=0.,vmax=1.)
tplot.axes.get_xaxis().set_visible(False)
tplot.axes.get_yaxis().set_visible(False)


def animate(i):
    chosentotem=totem[3*i]
    supertotem=np.zeros((3*Nl,4*Nl))
    supertotem[0:Nl,Nl:2*Nl]=chosentotem[4]
    supertotem[Nl:2*Nl,Nl:2*Nl]=chosentotem[1]
    supertotem[2*Nl:3*Nl,Nl:2*Nl]=chosentotem[5]
    supertotem[Nl:2*Nl,0:Nl]=chosentotem[0]
    supertotem[Nl:2*Nl,2*Nl:3*Nl]=chosentotem[2]
    supertotem[Nl:2*Nl,3*Nl:4*Nl]=chosentotem[3]
    tplot.set_array(supertotem)
    return None

animacion = animation.FuncAnimation(fig, animate, np.arange(1,Nt/3),interval=20, blit=False)
#animacion.save("/Users/juan/Desktop/cubic.mp4",fps=18,dpi=200)
plt.show()
```
